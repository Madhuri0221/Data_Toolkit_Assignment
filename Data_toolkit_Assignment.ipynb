```python
# 2) How does broadcasting work in NumPy?

"""
NumPy broadcasting allows operations on arrays of different shapes without
requiring explicit loops, making calculations more efficient

It automatically expands smaller arrays to match the shape of larger ones
when performing element-wise operations

When performing an operation between two arrays, NumPy checks their shapes:

 1) If the dimensions match exactly, element-wise operations proceed normally

 2) If dimensions differ, NumPy expands the smaller array along the missing
    dimensions, without copying data.

here: NumPy broadcasts the scalar 5 across the array, applying addition to
each element

import numpy as np
arr = np.array([1, 2, 3])
print(arr + 5)  # Every element is increased by 5

o/p: [6 7 8]

arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([10, 20, 30])
print(arr1 + arr2)

o/p: [[11 22 33]
 [14 25 36]]

"""
```

```python
# 17)  Explain the difference between apply() and map() in Pandas?

"""
Use map() for Element-wise Operations

import pandas as pd

df = pd.DataFrame({"Values": [1, 2, 3, 4]})
df["Squared"] = df["Values"].map(lambda x: x**2)
print(df)

Use apply() for Row/Column-wise Operations

import pandas as pd

# Sample DataFrame
df = pd.DataFrame({
    "Student": ["Alice", "Bob", "Charlie"],
    "Math": [85, 78, 92],
    "Science": [88, 76, 95],
    "English": [90, 82, 87]
})

# Applying row-wise sum to calculate total score
df["Total_Score"] = df.apply(lambda row: row["Math"] + row["Science"] +
row["English"], axis=1)

print(df)

o/p:
Student  Math  Science  English  Total_Score
0    Alice    85       88       90          263
1      Bob    78       76       82          236
2  Charlie    92       95       87          274

"""
```

```python
# 21) Why is NumPy’s array slicing faster than Python’s list slicing?

"""
NumPy’s array slicing is faster than Python’s list slicing due to efficient
memory handling and optimized internal implementation.

Key Reasons for Faster NumPy Slicing:
Contiguous Memory Layout → NumPy stores arrays in contiguous memory blocks,
unlike Python lists, which store elements as individual objects with references.

Views Instead of Copies → Slicing a NumPy array returns a view, not a copy,
reducing memory overhead, whereas Python lists create full copies.

C-Based Implementation → NumPy operations are highly optimized using C
and vectorized calculations, making slicing faster.

Fixed Data Types → NumPy arrays store elements of the same data type,
while Python lists hold mixed types, requiring additional processing.

"""
```

```python
# 4) How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [1, 3, 5, 3, 1],
    'D': [2, 2, 2, 2, 2]
})

corr_matrix = df.corr()
print(corr_matrix)

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix)

plt.title("Correlation Matrix Heatmap",pad=20)
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 9) Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold

import pandas as pd

d={"Colours":["Red","Blue","Yellow","Green"],
   "Pens":[1,2,3,4]}
df = pd.DataFrame(d)

print("Previous dataframe")
print()
print(df)

threshold = df['Pens'].median()
print()
print(f"Threshold (median): {threshold}")

df = df[df['Pens'] > threshold]
print()

print("Filtered rows")
print()
print(df)
```

```python
# 13) Create a 3D scatter plot using Plotly

import plotly.express as px
import pandas as pd
import numpy as np


np.random.seed(42)
df = pd.DataFrame({
    'X': np.random.randn(100),
    'Y': np.random.randn(100),
    'Z': np.random.randn(100),
    'Value': np.random.randint(1, 100, 100),
    'Category': np.random.choice(['A', 'B', 'C'], 100)
})

fig = px.scatter_3d(
    df,
    x='X',
    y='Y',
    z='Z',
    title='Basic 3D Scatter Plot',
    hover_name='Category',
    hover_data={'X': ':.2f', 'Y': ':.2f', 'Z': ':.2f'}
)


fig.update_layout(
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis',
        camera=dict(eye=dict(x=1.5, y=1.5, z=0.1))
    ),
    margin=dict(l=0, r=0, b=0, t=30),
    height=800
)

fig.show()
```
